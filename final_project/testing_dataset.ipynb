```python
# Bar plot over crime category frequency

arr_keys = data["Category"].value_counts().index.tolist()


arr_vals = data["Category"].value_counts().tolist()

import matplotlib.pyplot as plt 
fig = plt.figure(figsize = (10, 5))


plt.xticks(rotation = 'vertical')

plt.xlabel("Category")
plt.ylabel("No. of occurences")
plt.title("Crimes occured in SF")
plt.bar(arr_keys, arr_vals)
```
